```python
plt.subplot(121) #1- liczba wierszy , 2- liczba kolumn , 1-odnosi si do konkretnego wykresu: skoro 1 to odnosi si do pierwszego wykresu od lewej
```

```python
plt.subplot(121) # dane poniej s dla pierwszego wykresu od lewej
plt.plot([2,3,1] , [0,5,7])
plt.subplot(122) # dane poniej s dla drugiego wykresu od lewej
plt.plot([0,3,2] , [3,5,1] , 'r--') # r-- oznacza, e linia bdzie przerywana
```

```python
plt.subplot(131) # 3 oznacza e bd 3 wykresy
plt.plot([2,3,1] , [0,5,7])
plt.subplot(132)
plt.plot([0,3,2] , [3,5,1] , 'r--')
plt.subplot(133)
plt.plot([0,2,2] , [7,5,1] , 'ro') # ro czyli wykres punktowy
```

```python
fig = plt.figure(figsize=(14,14)) # zwikszamy rozmiar wykresu
plt.subplot(131)
plt.plot([2,3,1] , [0,5,7])
plt.subplot(132)
plt.plot([0,3,2] , [3,5,1] , 'r--')
plt.subplot(133)
plt.plot([0,2,2] , [7,5,1] , 'ro')

fig.show() # trzeba to zapisa np. w programie Spyder, inaczej nie wywietli nam obiektu (bo zdefiniowalimy nowy obiekt fig podczas zmiany jego wielkoci)
```

```python
plt.barh(y=['a' , 'b' , 'c' , 'd'] , width=[1.2, 3.2, 1.2, 5.2])
```

```python
x1 = np.random.randn(300)
x2 = np.random.randn(300)
plt.scatter(x=x1 , y=x2)
```
